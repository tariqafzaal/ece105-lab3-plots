```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=8986)

sensor_a = rng.normal(loc=25, scale=3, size=200)
sensor_b = rng.normal(loc=27, scale=4.5, size=200)
timestamps = rng.uniform(0.0, 10.0, size=200)

print("sensor_a.shape:", sensor_a.shape, "dtype:", sensor_a.dtype)
print("sensor_b.shape:", sensor_b.shape, "dtype:", sensor_b.dtype)
print("timestamps.shape:", timestamps.shape, "dtype:", timestamps.dtype)
```
